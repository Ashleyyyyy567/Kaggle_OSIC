### 1.3 Examination of Correlation

```python
# examine the correlation between FVC and Weeks, separated by Sex
fig = px.scatter(train, x="Weeks", y="FVC", color='Sex',title = 'Coorelation between FVC and Weeks')
fig.show()
```

It seems like Male over the course of Weeks has generally greater FVC comparing to that of Female. Interested in whether Male and Female FVC are drawn from the same population, it uses permutation test to examine if such pattern is random and due to chance, or there is indeed a difference between `FVC` and `Sex`
- H0: Male and Female FVC are drawn from the same population
- Ha: Male and Female FVC are drawn from the different population
- Test statistics: 0.05

```python
n = 1000
train_copy = train.copy()
differences = []
for _ in range(n):
    
    # shuffle the sex
    index = train_copy['Sex'].sample(replace=False, frac=1).reset_index(drop=True)
    
    # put them in a table
    shuffled = train_copy.assign(**{'Shuffled Sex': index})
    
    # compute the group differences
    group_means = shuffled.groupby('Shuffled Sex').mean().loc[:, 'FVC']
    difference = group_means.diff().iloc[-1]
    
    # add it to the list of results
    differences.append(difference)
```

```python
obs_mean = train.groupby('Sex').mean().loc[:, 'FVC']
obs_diff = obs_mean.diff().iloc[-1]
obs_diff
```

```python
diff_df = pd.DataFrame(differences)
diff_df = diff_df.rename(columns={0:'Differences'})
fig = px.histogram(diff_df, x="Differences")
fig.add_trace(go.Scatter(mode='markers',x=[obs_diff],y=[0],showlegend=False))
fig.show()
```

```python
# p value calculation
p_value = (differences >= obs_diff).mean()
p_value
```

Since p-value is 0 which is smaller than 0.05, the test reject N0 and claim that FVC of difference sex are drawn from different populaiton. It is test significant.

## 2. Construct model

```python
def data_cleaning(train): 
    X = train.drop(['Patient','Percent'], axis = 1)
    
    # encode str into label
    label = LabelEncoder() 
    X['Sex'] = label.fit_transform(X['Sex']) 
    X['SmokingStatus'] = label.fit_transform(X['SmokingStatus']) 

    # apply one hot encoder
    onehotencoder = OneHotEncoder() 
    columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [3, 4])], remainder='passthrough') 
    train_X = np.array(columnTransformer.fit_transform(X), dtype = np.float32)

    # convert numpy array to torch tensor
    X_tensor = torch.from_numpy(train_X)
    y_tensor = torch.tensor(train['Percent'].values.astype(np.float32))
    return X_tensor, y_tensor

X_tensor,y_tensor = data_cleaning(train)
```

```python
# define the linear network with one hidden layer
class Net(torch.nn.Module): 
    def __init__(self, n_input, n_hidden, n_output): 
        super(Net, self).__init__()
        self.hidden = nn.Linear(n_input, n_hidden)
        self.output = nn.Linear(n_hidden, n_output)
        
    def forward(self, x): 
        x = F.relu(self.hidden(x))     
        output = self.output(x)            
        return output
net = Net(8, 15, 1)
print(net)
```

```python
# define the loss function 
loss_function = nn.MSELoss()
optimizer = torch.optim.SGD(net.parameters(), lr=0.2)
epoch = 500
for i in range(epoch): 
    output = net(X_tensor)
    loss = loss_function(output, y_tensor)
    optimizer.zero_grad()  
    loss.backward()       
    optimizer.step() 
```

```python
def clean_test(train): 
    X = train.drop(['Patient','Percent'], axis = 1)
    X['Male'] = (X['Sex'] != 'Male').astype(int)
    X['Female'] = (X['Sex'] == 'Male').astype(int)
    X['Currently smokes'] = (X['SmokingStatus'] == 'Currently smokes').astype(int)
    X['Ex-smoker'] = (X['SmokingStatus'] == 'Ex-smoker').astype(int)
    X['Never smoked'] = (X['SmokingStatus'] == 'Never smoked').astype(int)
    X = X[['Male', 'Female', 'Currently smokes','Ex-smoker','Never smoked', 'Weeks', 'FVC', 'Age']]

    X_test = torch.tensor(X.values.astype(np.float32))
    y_test = torch.tensor(train['Percent'].astype(np.float32))
    return X_test, y_test

X_test, y_test = clean_test(test)
```
